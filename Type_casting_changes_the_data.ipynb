```python
from type_casting import Type_casting       
from IPython.display import display
import pandas as pd
import os
files = list(map(str.strip, os.popen('find /home/amir/github/LFD_projects_3 -name "*.csv"')) )
for i in files: 
    n = 0
    # i = i.split(",")[1].lstrip("/")
    try:
        df = pd.read_csv(i) 
    except:
        pass
    df_copy = df.copy()
    Type_casting(id(df)) 
    cols = df.eq(df_copy).mean().where(lambda x:x<1).dropna().index.to_list()
    if not cols:
        continue
    for diff_col in cols:
        diff_df = (
            pd.DataFrame({"before" : df_copy[diff_col], "after" : df[diff_col]})
            .dropna()
            .where(lambda x:x.before != x.after)
            .dropna()
        )
        if not len(diff_df):
            continue
        if df_copy[diff_col].dtype in ['int64', 'float64']:
            diff_df['difference'] = diff_df.after.sub(diff_df.before)
            diff_df = diff_df.astype(int).where(lambda x:x.difference != 0).dropna()
            if not(len(diff_df)):
                continue
        if n == 0:
            print("\n\n\n>>>>>>>>>>\tFile:", i)
            print(df.shape)
            n += 1

        print(f"Different count: {diff_df.shape[0]},\nDifferent ratio(1-100): {round( len(diff_df)/len(df)*100)}")
        print("==============\tCol:", diff_col)
        display(diff_df)
        print("\n\n\n\n")
```
