```python
import pandas as pd
df = pd.read_csv("sample_amir.csv")
if "Unnamed: 0" in df.columns:
    df = df.drop("Unnamed: 0", axis=1)
```

```python
# ACCESS_METHOD_ID varibale me aksar values 2 dafa repeat hwi hen, magar kuch values 1 dafa i hen, jo 1 dafa hen un k lye 2nd line
# add karni h, jis me access_method_id call to same ho, or IsWeekDay me oposite ho(agar exist me ho to new me 1 and vise versa)
# or baqi columns me 0
def add_row(row):
    global df
    df.loc[-1] = row
    df.index = df.index + 1  
    
counts = df.ACCESS_METHOD_ID_.value_counts()
for i in counts[counts == 1].index:
    m = 0 if df[df.ACCESS_METHOD_ID_ == i].IsWeekDay_.values == 1 else 1
    add_row([i, m, 0, 0, 0, 0, 0, 0, 0,  0])
```

# Two rows to one row

```python
new_columns = ['ACCESS_METHOD_ID', 
'IsWeekDay_0',
'CALL_DURATION_sum_0',
'CALL_DURATION_max_0', 
'CALL_DURATION_min_0', 
'CALL_DURATION_mean_0',
'NO_OF_CALLS_sum_0', 
'NO_OF_CALLS_max_0', 
'NO_OF_CALLS_min_0',
'NO_OF_CALLS_mean_0',
'IsWeekDay_1',
'CALL_DURATION_sum_1',
'CALL_DURATION_max_1', 
'CALL_DURATION_min_1', 
'CALL_DURATION_mean_1',
'NO_OF_CALLS_sum_1', 
'NO_OF_CALLS_max_1', 
'NO_OF_CALLS_min_1',
'NO_OF_CALLS_mean_1']
adf = pd.DataFrame(columns=new_columns)
df = df.sort_index()
for i in df.ACCESS_METHOD_ID_.unique():
    bdf = df[df.ACCESS_METHOD_ID_ == i]
    l = [i]
    for z in df.columns[1:]:
        l.append(bdf[z].values[0])
        l.append(bdf[z].values[1])
    adf.loc[-1] = pd.Series(l, new_columns)
    adf.index = adf.index + 1  
```
